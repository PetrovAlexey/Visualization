Основная фунция построения дерева


```
def draw(vertex, depth):
  local_coord = dict()

  if len(nodes_edges[vertex]) == 0:
    contour = np.array([[0, 0]])
    local_coord[vertex] = (0, depth)

    return local_coord, contour
  
  if len(nodes_edges[vertex]) == 1:
    local_coord, left_contour = draw(nodes_edges[vertex][0], (depth-1))
    contour = np.array([ np.array([0, 0]), *left_contour])
    local_coord[vertex] = (0, depth)
    return local_coord, contour

  if len(nodes_edges[vertex]) == 2:
    local_coord, left_contour = draw(nodes_edges[vertex][0], (depth-1))
    
    right_coord, right_contour = draw(nodes_edges[vertex][1], (depth-1))

    min_depth = min(len(left_contour), len(right_contour))
    max_depth = max(len(left_contour), len(right_contour))

    rev_left = left_contour
    rev_right = right_contour

    # find shift
    shift = 0
    for i in range(min_depth):
      temp = rev_left[i][1]  + 1 - rev_right[i][0]
      if temp > shift:
        shift = temp

    for i in right_contour:
      i[1] += shift
      i[0] += shift

    rev_left = left_contour
    rev_right = right_contour
    
    contour = np.array(([[0, 0] for i in range(max_depth)]))
    for i in range(max_depth):
      if i >= len(left_contour):
        left = rev_right[i][0]
        right = rev_right[i][1]
      elif i >= len(right_contour):
        left = rev_left[i][0]
        right = rev_left[i][1]
      else:
        left = min(rev_left[i][0], rev_right[i][0])
        right = max(rev_left[i][1], rev_right[i][1])

      contour[i][0] = left
      contour[i][1] = right
    
    root = 0
    local_coord[vertex] = (root, depth)
    
    contour =  np.array([np.array([root, root]), *contour])

    shift_right_coord = dict()
    for (node, coord) in right_coord.items():
      shift_right_coord[node] = (coord[0]+shift, coord[1])
    
    local_coord.update(shift_right_coord)
    return local_coord, contour
  
```

Парсим дерево из формата Graphml и ищем корень дерева



```
parser = GraphMLParser()
g = parser.parse('tree1.xml')
nodes_set = set()
nodes_edges = dict()
nodes_coord = dict()

nodes = g.nodes()
for node in nodes:
    nodes_set.add(node.id)
    nodes_edges[node.id] = np.empty(0)
    nodes_coord[node.id] = np.zeros(2)

edges = g.edges()
for edge in edges:

    nodes_edges[edge.node1.id] = np.append(nodes_edges[edge.node1.id], edge.node2.id)
    nodes_set.remove(edge.node2.id)
```

```
test, contour_test = draw(list(nodes_set)[0], 0)
```

```
print(contour_test)
```

Отрисуем дерево


```
import matplotlib.ticker as ticker
fig, ax = plt.subplots()

ax.plot([i[0] for i in test.values()], [i[1] for i in test.values()], 'o', color='black');

for (node, children) in nodes_edges.items():
  parent_coord = test[node]
  for child in children:
    child_coord = test[child]
    x_values = (parent_coord[0], child_coord[0])
    y_values = (parent_coord[1], child_coord[1])
    ax.plot(x_values, y_values, color='green')

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(2))

ax.grid("True")
fig.set_figwidth(12)
fig.set_figheight(8)


plt.show()
```
